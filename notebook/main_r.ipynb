data normalization

it is a way to bring all data into a similar range, for more meaningful comparison.

```r
range(df$ArrDelay)
```

```r
head(df$ArrDelay)
```

```r
simple_scale <- df$ArrDelay / max(df$ArrDelay)
print(head(simple_scale))
```

```r
minmax_scale <- (df$ArrDelay - min(df$ArrDelay)) / (max(df$ArrDelay) - min(df$ArrDelay))
print(head(minmax_scale))
```

```r
# Z-score normalization
z_scale <- (df$ArrDelay - mean(df$ArrDelay)) / sd(df$ArrDelay)
print(head(z_scale))
```

Binning: groups a set of numerical values into a set of bins.

```r
pl <- ggplot(data = df, mapping = aes(x = ArrDelay)) + 
geom_histogram(bins = 100, color = "white", fill = "red") +
coord_cartesian(xlim = c(-73, 682))

print(pl)
```

```r
binning <- df %>% mutate(quartile_rank = ntile(df$ArrDelay, 4)) 
print(binning)
```

binning improves the accuracy of the model.
helps to better understand the data.

quartiles refer to the boundaries that divide observations into four defined intervals.
